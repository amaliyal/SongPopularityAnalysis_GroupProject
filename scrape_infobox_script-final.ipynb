```python
import pandas as pd
import wptools as wp
import json
#from timeit import default_timer as timer
```

```python
music = pd.read_csv('spotify track artist audio.csv')
music                  
```

```python
# function for scraping singer_details and saving as a JSON dictionary to a text file
def scrape_infobox(file_name, artist_name):
    count = 0 
    fail_dict = {}
    with open(file_name, 'w', encoding='utf-8') as outfile:
        for each_name in artist_name:
            count += 1
            print(str(count) + ": " + each_name)
            try:
                page = wp.page(each_name).get_parse()
                json.dump(page.data['infobox'], outfile)
                outfile.write('\n')
            except LookupError as err1:
                fail_dict[each_name] = err1
                print('lookup error')
                pass
            except ValueError as err2:
                fail_dict[each_name] = err2
                print('value error')
                pass
        
        return
```

```python
unique_artist_names = music['artist_name'].unique()
len(unique_artist_names)
```

```python
# 856 artists
artist_names_1 = unique_artist_names[:800]

# 800 artists
artist_names_2 = music['artist_name'].unique()[800:1600]

# 642 artists
artist_names_3 = music['artist_name'].unique()[1600:2400]

# 642 artists
artist_names_4 = music['artist_name'].unique()[2400:]
```

```python
# First 800 unique Artiste
scrape_infobox(file_name='singers_json_1.txt', artist_name=artist_names_1)
```
