Gather name, age, origin

```python
import json
import numpy as np
import pandas as pd
```

```python
def create_df(file_name):
    singers_details = []
    with open (file_name, encoding = 'utf-8') as file:
        for each_infobox in file.readlines():
            each_line = json.loads(each_infobox)
            keys = ['name', 'members', 'current_members', 'birth_name', 'origin', 'birth_place', 'birth_date']
            try:
                if keys[0] in each_line.keys():  # artist name
                    name = each_line[keys[0]]
                else:
                    name = np.nan

                if keys[1] in each_line.keys():  # members for bands
                    members = each_line[keys[1]]
                elif keys[2] in each_line.keys():
                    members = each_line[keys[2]]
                else:
                    members = np.nan

                if keys[3] in each_line.keys():  # artist birth name
                    birth_name = each_line[keys[3]]
                else:
                    birth_name = np.nan

                if keys[4] in each_line.keys():  # nationality
                    birth_place = each_line[keys[4]]
                elif keys[5] in each_line.keys():
                    birth_place = each_line[keys[5]]
                else:
                    birth_place = np.nan

                if keys[6] in each_line.keys():  # age
                    birth_date = each_line[keys[6]]
                else:
                    birth_date = np.nan

            except AttributeError as err:
                pass

            singers_details.append({'names': name,
                                    'members': members,
                                    'birth_name': birth_name, 
                                    'birth_place': birth_place,
                                    'birth_date': birth_date})
            
    artists_details = pd.DataFrame(singers_details)
           
    return artists_details
```

```python
df1 = create_df(file_name='singers_json_1.txt')
df2 = create_df(file_name='singers_json_2.txt')
df3 = create_df(file_name='singers_json_3.txt')
df4 = create_df(file_name='singers_json_4.txt')
```

```python
artists = df1.append([df2, df3, df4], ignore_index=True)
artists = artists.drop_duplicates(keep='first')
artists
```

```python
gender_df = pd.read_csv('singers_gender.csv', encoding='latin1')
gender_df
```

```python
artists_details = artists.merge(gender_df, how='left', left_on='names', right_on='artist')
artists_details
```

```python
artists_details.to_csv('artists_details.csv', index=False)
```
